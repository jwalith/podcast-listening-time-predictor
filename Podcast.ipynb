```python
df = pd.read_csv("train.csv")
```

```python
df = df.head(5000)
```

```python
df = df.drop(columns=['id', 'Podcast_Name', 'Episode_Title'])
```

```python
X_train =  df.drop(columns = ['Listening_Time_minutes'])
```

```python
y_train = df['Listening_Time_minutes']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
```

```python
categorical_cols = ['Genre', 'Publication_Day', 'Publication_Time', 'Episode_Sentiment']
```

```python
numerical_cols = [c for c in X_train.columns if c not in categorical_cols]
```

```python
numeric = SimpleImputer(strategy = 'median')
```

```python
categorical = OneHotEncoder(handle_unknown = 'ignore')
```

```python
# ColumnTransformer is a tool from scikit-learn that lets you: 
# Apply different preprocessing steps to different columns at the same time.

preprocessor = ColumnTransformer(
                transformers = [
                        ('n',numeric,numerical_cols),
                        ('c',categorical,categorical_cols)
                ]
    
)
```

```python
model_pipe = Pipeline( steps = [
            ('preprocessor',preprocessor),
            ('model', RandomForestRegressor(
    n_estimators=75,  # 100 trees
    random_state=42,
    n_jobs=-1          # Use all cores
))
]
)
```

```python
model_pipe.fit(X_train, y_train)
```

```python
import joblib

joblib.dump(model_pipe, 'final_model_pipeline.pkl')
```
